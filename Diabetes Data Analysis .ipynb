## Questions for Exploratory Data Analysis

##### 1. What is the distribution Of the Outcome variable (0 vs 1)?

##### 2. What is the percentage of diabetic vs non-diabetic cases?

```python
# Count the number of occurrences for each outcome (0 and 1)
outcome_distribution = df['Outcome'].value_counts()

# Calculate the total number of cases
total_cases = len(df)

# Calculate the percentage of non-diabetic (Outcome = 0) and diabetic (Outcome = 1)
percentage_non_diabetic = (outcome_distribution[0] / total_cases) * 100
percentage_diabetic = (outcome_distribution[1] / total_cases) * 100

# Print the results
print(f"Percentage of non-diabetic cases: {percentage_non_diabetic:.2f}%")
print(f"Percentage of diabetic cases: {percentage_diabetic:.2f}%")
```

##### 4. Are there any outliers in the features?

```python
# Function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Check for outliers in each feature (excluding 'Outcome')
for column in df.columns[:-1]:  # Exclude 'Outcome'
    outliers = detect_outliers(df, column)
    print(f"Number of outliers in {column}: {len(outliers)}")

    if len(outliers) > 0:
        print(f"Outliers in {column}:\n", outliers[column].values)
    print('-' * 50)
```

```python
# Visualize outliers using boxplots
df.drop(columns='Outcome').boxplot(figsize=(8, 6), grid=False, rot=45)
plt.title('Boxplot for Each Feature')
plt.show()
```

##### 6. Which features are most strongly correlated with the Outcome?

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Extract the correlation values for 'Outcome'
outcome_correlation = correlation_matrix['Outcome'].drop('Outcome')  # Exclude 'Outcome' itself

# Sort the correlations in descending order to find the strongest correlations
outcome_correlation_sorted = outcome_correlation.sort_values(ascending=False)

# Display the correlations with 'Outcome'
print("Features most strongly correlated with 'Outcome':\n", outcome_correlation_sorted)
```

##### 7. How does the age distribution differ between diabetic and non-diabetic individuals?

```python
import matplotlib.pyplot as plt

# Filter the dataset into diabetic and non-diabetic groups
diabetic = df[df['Outcome'] == 1]
non_diabetic = df[df['Outcome'] == 0]

# Plot histograms to compare age distribution
plt.figure(figsize=(8, 4))

plt.hist(non_diabetic['Age'], bins=20, alpha=0.5, label='Non-Diabetic', color='blue')
plt.hist(diabetic['Age'], bins=20, alpha=0.5, label='Diabetic', color='red')

plt.title('Age Distribution: Diabetic vs Non-Diabetic Individuals')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Plot KDE (density) plots for age distribution
plt.figure(figsize=(8, 4))

sns.kdeplot(non_diabetic['Age'], fill=True, label='Non-Diabetic', color='blue')
sns.kdeplot(diabetic['Age'], fill=True, label='Diabetic', color='red')

plt.title('Age Distribution: Diabetic vs Non-Diabetic Individuals')
plt.xlabel('Age')
plt.legend()
plt.grid(True)
plt.show()
```

##### 8. Is there a specific age group that is more prone to diabetest?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create age bins (e.g., 0-20, 21-30, 31-40, etc.)
bins = [0, 20, 30, 40, 50, 60, 70, 80]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80']

# Create a new column 'AgeGroup' to categorize ages into bins
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculate the proportion of diabetic cases (Outcome = 1) in each age group
age_group_diabetes = df.groupby('AgeGroup', observed=True)['Outcome'].mean() * 100



# Plot the results
plt.figure(figsize=(10, 6))
age_group_diabetes.plot(kind='bar', color='skyblue')
plt.title('Percentage of Diabetic Individuals by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Diabetic Cases (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

##### 9. How do glucose levels differ between diabetic and non-diabetic individuals?

```python
import matplotlib.pyplot as plt

# Split the data into diabetic and non-diabetic groups
diabetic = df[df['Outcome'] == 1]
non_diabetic = df[df['Outcome'] == 0]

# Plot histograms to compare glucose level distribution
plt.figure(figsize=(10, 6))

plt.hist(non_diabetic['Glucose'], bins=20, alpha=0.5, label='Non-Diabetic', color='blue')
plt.hist(diabetic['Glucose'], bins=20, alpha=0.5, label='Diabetic', color='red')

plt.title('Glucose Level Distribution: Diabetic vs Non-Diabetic Individuals')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot boxplots to compare glucose levels
plt.figure(figsize=(10, 8))
plt.boxplot([non_diabetic['Glucose'], diabetic['Glucose']], labels=['Non-Diabetic', 'Diabetic'])
plt.title('Glucose Level Comparison: Diabetic vs Non-Diabetic Individuals')
plt.ylabel('Glucose Level')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Plot KDE (density) plots for glucose levels
plt.figure(figsize=(10, 6))

sns.kdeplot(non_diabetic['Glucose'], fill=True, label='Non-Diabetic', color='blue')
sns.kdeplot(diabetic['Glucose'], fill=True, label='Diabetic', color='red')

plt.title('Glucose Level Distribution: Diabetic vs Non-Diabetic Individuals')
plt.xlabel('Glucose Level')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Summary statistics for glucose levels in both groups
diabetic_glucose_mean = diabetic['Glucose'].mean()
non_diabetic_glucose_mean = non_diabetic['Glucose'].mean()

print(f"Average Glucose Level - Diabetic: {diabetic_glucose_mean:.2f}")
print(f"Average Glucose Level - Non-Diabetic: {non_diabetic_glucose_mean:.2f}")
```
